```python
# 下記は2secを超えてしまう。　頓挫中

import re

input_S = input()

char_list = ["dream", "dreamer", "erase", "eraser"]
char_list.sort(reverse=True, key=len)

is_found =False
is_YES =False

for j in range(10000):
    is_found = False
    for i in char_list:
        if input_S.endswith(i):
            re_word = i + "$"
            input_S = re.sub(re_word,"",input_S)
#             print(input_S)
            is_found = True
            if len(input_S) == 0:
                is_YES =True
                print("YES")
                break
            else:
#                 print("continue")
                continue
    else:
        if is_found is False:
#             print("break")
            break
        continue
    break

if is_YES is False:
    print("NO")
```
